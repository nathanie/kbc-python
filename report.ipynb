```python
from clients import client
import pickle
f = open('client_list.pickle','rb')
clients = pickle.load(f)
f.close()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots(1,2,figsize=(15,4))
_ = ax[0].hist([x.get_balance_without_comm_fee() for x in clients],bins=40)
ax[0].set_title('histogram of balance before commission fee')
ax[0].set_xlabel('balance')
ax[0].set_ylabel('number of customers')
_ = ax[1].hist([x.get_balance_after_comm_fee() for x in clients],bins=40)
ax[1].set_title('histogram of balance after commission fee')
ax[1].set_xlabel('balance')
ax[1].set_ylabel('number of customers')
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,4))
_ = ax.hist([x.get_comm_fee() for x in clients],bins=20)
ax.set_title('histogram of transactions number made')
ax.set_xlabel('# transactions')
ax.set_ylabel('number of customers')
```
